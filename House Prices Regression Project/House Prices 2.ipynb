```python
#Importing necessary libaries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"train.csv")
```

```python
df.head()
```

```python
n = dict(df.isnull().sum())
```

```python
n
```

```python
df.info()
```

```python
multiplied_dict = {k: v / 1460 * 100 for k, v in n.items()}
multiplied_dict
```

```python
colu = {k: v for k, v in multiplied_dict.items() if v > 45}
```

```python
drop_col = colu.keys()
```

```python
df.drop(columns= drop_col , inplace= True)
```

```python
df.head()
```

```python
df.info()
```

```python
fig_ = df.hist(figsize=(25, 30), bins=50, color="darkcyan", edgecolor="black", xlabelsize=8, ylabelsize=8)
#Quick look at outliers
```

```python
df.info()
```

```python
cols_with_missing = df.columns[df.isnull().any()].tolist()
```

```python
  

for lol in cols_with_missing:
    df[lol] = df[lol].fillna(df[lol].mode().iloc[0])
```

```python
df.info()
```

```python
categorical = df.select_dtypes(include='object')
```

```python
from sklearn.preprocessing import LabelEncoder
for i in categorical.columns:
    label_encoder = LabelEncoder()
    df[i] = label_encoder.fit_transform(df[i])

df.info()
```

```python
plt.subplots(figsize=(50,50))
sns.heatmap(df.corr(),  vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot_kws={"size": 9, "color": "black"}, square=True, annot=True)
```

```python
corrr = dict(df.corrwith(df['SalePrice']))
```

```python
corrr
```

```python
droppin = {k: v for k, v in corrr.items() if v >= 0.25 or v <= -0.25}
```

```python
droppin
colls_tokeep = droppin.keys()
```

```python
colls_tokeep = list(colls_tokeep)
colls_tokeep
```

```python
dro = df.drop(columns=colls_tokeep)
```

```python
df.drop(columns=dro.columns, inplace=True)
```

```python
df
```

```python
y = df['SalePrice']
X = df.drop(columns='SalePrice')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2023)
```

```python
leg = LinearRegression()
leg.fit(X_train, y_train)
y_pred = leg.predict(X_test)
leg.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error as MSE
score = MSE(y_test, y_pred)
```
